```python
# list
my_list = [5,6,3,8,9,2,1,0,5]

def oracle(input):
  winner = 1
  if input is winner:
    response = True
  else:
    response = False

  return response


for i, number in enumerate(my_list):
  if oracle(number) is True:
    print(i+1)
    print("counts made")
```

```python
from qiskit import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the oracle circuit

# Create a new circuit.

# A circuit is a list of instructions bound to some registers.

oracle = QuantumCircuit(2, name="oracle")
# regs (list(Register) or list(int)) – The registers to be included in the circuit.

oracle.cz(0,1)
oracle.to_gate()
oracle.draw()
```

```python
backend = Aer.get_backend('statevector_simulator')
grovercirc = QuantumCircuit(2,2)
grovercirc.h([0,1])
grovercirc.append(oracle,[0,1])
grovercirc.draw()
```
